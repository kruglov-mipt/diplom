```python
import pyons
from examples.roadrfidsim.roadrfidsim.journal import Journal
from examples.roadrfidsim.roadrfidsim.model import Model, Factory, \
    ModelDescriptor
    
@pyons.stop_condition()
def check_sim_time():
    return pyons.time() > 2.0
```

```python
pyons.reset()
md = ModelDescriptor()
md.lanes_number = 1
md.vehicle_tag_locations = ['front']
md.reader_antennas_sides = ['front']
md.vehicle_length = 0.0
md.tag_start_offset = 20
factory = Factory(md)

model = Model(md)
model.channel = factory.build_channel()
model.reader = factory.build_reader(model.channel)
model.add_vehicle(factory.build_vehicle(0, model.channel))
# model.add_vehicle(factory.build_vehicle(1, model.channel))

pyons.set_model(model)

pyons.setup_env(log_level=pyons.LogLevel.WARNING)
pyons.run(max_events=1e6)

# Journal().print_all()
tag_rx_power = None
print("tag_rx_power: {}".format(tag_rx_power))
tag_rx_power = Journal().tag_rx_power_data_frame
```
